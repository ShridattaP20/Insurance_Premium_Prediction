```python
df['sex']=df.sex.map({'male':1,'female':0})
```

```python
df['smoker']=df.smoker.map({'yes':1,'no':0})
```

```python
df['region']=df.region.map({'northeast':1,'northwest':2,'southeast':3,'southwest':4})
```

```python
X=df.drop('expenses',axis=1)
y=df['expenses']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score,mean_squared_error
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
```

```python
sc=StandardScaler()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
X_train=sc.fit_transform(X_train)
```

```python
X_test=sc.transform(X_test)
```

```python
lr=LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pre=lr.predict(X_test)
```

```python
r2_score(y_test,y_pre)
```

```python
np.sqrt(mean_squared_error(y_test,y_pre))
```

```python
rf=RandomForestRegressor()
```

```python
rf.fit(X_train,y_train)
```

```python
rf_pre=rf.predict(X_test)
```

```python
r2_score(y_test,rf_pre)
```

```python
np.sqrt(mean_squared_error(y_test,rf_pre))
```

```python
xgb=XGBRegressor()
```

```python
xgb.fit(X_train,y_train)
```

```python
xg_pre=xgb.predict(X_test)
```

```python
r2_score(y_test,xg_pre)
```

```python
np.sqrt(mean_squared_error(y_test,xg_pre))
```
